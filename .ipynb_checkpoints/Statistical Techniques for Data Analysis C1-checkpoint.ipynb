```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
n = 10  
p = 0.7  
```

```python
x = np.arange(0, n + 1) 
probabilities = binom.pmf(k=x, n=n, p=p) 
```

```python
plt.figure(figsize=(10, 5))
plt.bar(x, probabilities, color='blue', width=0.4)
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('Probability distribution of a binomial random variable')
plt.show()
```

```python
quantiles_teoricos = sm.ProbPlot(probabilities).theoretical_quantiles
price_sorted = np.sort(probabilities)
plt.figure(figsize=(8, 6))
plt.scatter(quantiles_teoricos, price_sorted, color='blue', alpha=0.5, label='Q-Q plot')
plt.plot([np.min(quantiles_teoricos), np.max(quantiles_teoricos)], [np.min(price_sorted), np.max(price_sorted)], color='red', label='LÃ­nea de ajuste')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Observed quantiles')
plt.title('Q-Q plot of binomial distribution')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = data3.drop('price', axis=1)
y = data3['price']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()
```

```python
X_imputed = imputer.fit_transform(X)
X_scaled = scaler.fit_transform(X_imputed)
```

```python
train_sizes = [0.2, 0.25, 0.3]
train_size_names = [f"Train Size {int(size * 100)}%" for size in train_sizes]
accuracies = []
```

```python
for train_size in train_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=1 - train_size, random_state=42)
    print(train_size_names)
```

```python
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=test_size, random_state=42)
```

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
accuracies = [] 

for train_size in train_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=1 - train_size, random_state=42)
    classifier = GaussianNB()
    classifier.fit(X_train, y_train)
    accuracy = cross_val_score(classifier, X_train, y_train, cv=5).mean()
    accuracies.append(accuracy)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, accuracies, marker='o')
plt.title('Model accuracy vs. Training set size')
plt.xlabel('Training set size')
plt.ylabel('Cross-validated accuracy')
plt.xticks(train_sizes, labels=[f'{int(size * 100)}%' for size in train_sizes])
plt.grid(True)
plt.show()
```

Regression Line

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming train_sizes and accuracies are lists or numpy arrays of numbers
train_sizes = np.array(train_sizes)  
accuracies = np.array(accuracies)

# Calculate the regression line
slope, intercept = np.polyfit(train_sizes, accuracies, 1)
regression_line = slope * train_sizes + intercept

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(train_sizes, accuracies, marker='o')

# Add the regression line
plt.plot(train_sizes, regression_line, color='red', label='Regression Line')

plt.title('Model accuracy vs. Training set size')
plt.xlabel('Training set size')
plt.ylabel('Cross-validated accuracy')


plt.xticks(train_sizes, labels=[f'{int(size * 100)}%' for size in train_sizes])

plt.grid(True)
plt.legend()
plt.show()
```
