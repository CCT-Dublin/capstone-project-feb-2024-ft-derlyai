# PCA to reduce noise in data

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
np.random.seed(0)
n_samples = 1000
n_features = 10
noise = np.random.normal(loc=0, scale=0.5, size=(n_samples, n_features))
data = np.random.normal(loc=0, scale=1, size=(n_samples, n_features)) + noise
data3 = pd.DataFrame(data, columns=[f'Feature_{i}' for i in range(1, n_features + 1)])
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data3)
```

```python
n_components = 5  
pca = PCA(n_components=n_components)
pca.fit(scaled_data)
```

```python
principal_components = pca.components_
```

```python
reconstructed_data = np.dot(pca.transform(scaled_data), principal_components) + scaler.mean_
```

```python
reconstructed_data3 = pd.DataFrame(reconstructed_data, columns=data3.columns)
```

```python
print("new.csv:")
print(data3.head())
print("\nReconstructed data:")
print(reconstructed_data3.head())
```
